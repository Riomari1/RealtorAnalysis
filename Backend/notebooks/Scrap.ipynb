```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from selenium.webdriver.firefox.service import Service
import time

service = Service()
options = webdriver.FirefoxOptions()
driver = webdriver.Firefox(service=service, options=options)

url = 'https://www.propertyfinder.ae/en/search?c=2&fu=0&rp=y&ob=mr'

# Open the URL using Selenium
driver.get(url)

# Wait for the page to fully load (you may need to adjust the waiting time as needed)
wait = WebDriverWait(driver, 10)

def scroll_to_element(element):
    driver.execute_script("arguments[0].scrollIntoView();", element)
    time.sleep(15)  # Optional: Add a short delay after scrolling to stabilize the page

while True:
    try:
        # Wait for the next button to become clickable
        next_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '[data-testid="pagination-page-next-button"]')))
                
        # Get the page source after it has fully loaded
        page_source = driver.page_source
                
        # Parse the page source with BeautifulSoup
        soup = BeautifulSoup(page_source, 'html.parser')
        property_listings = soup.find_all(role="listitem")

        # Loop through each property listing
        for listing in property_listings:
            try:
                # Extract details for each property
                propertyId = listing.get('data-id')  
                type = listing.find('p', {'data-testid': 'property-card-type'}).text.strip()
                price = listing.find('p', {'data-testid': 'property-card-price'}).text.strip()
                bed = listing.find('p', {'data-testid': 'property-card-spec-bedroom'}).text.strip()
                bath = listing.find('p', {'data-testid': 'property-card-spec-bathroom'}).text.strip()
                area = listing.find('p', {'data-testid': 'property-card-spec-area'}).text.strip()
                location = listing.find('div', {'data-testid': 'property-card-location'}).text.strip()
                date = listing.find('p', {'class': 'styles-module_footer__publish-info__UVabq'}).text.strip()

                # Print extracted details for each property
                print(f"Property Id: {propertyId}")
                print(f"Property Type: {type}")
                print(f"Property Bedroom: {bed}")
                print(f"Property Bathroom: {bath}")
                print(f"Property Area: {area} sqft")
                print(f"Property Price: {price}")
                print(f"Property Location: {location}")
                print("----------------------------------------")  # Separator between listings
            except Exception as e:
                print(f"Error extracting details: {e}")
        
        scroll_to_element(next_button)

        
        # Click the next button to move to the next page
        next_button.click()
        
        # Increment page number        
    except Exception as e:
        print(f"No more pages to scrape. Error: {e}")
        break

# Close the Selenium WebDriver
driver.quit()
```
